```python
data = df.drop('Name', axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
#giving numerical values to legendary column
data.iloc[:, 5:6] = data.iloc[:, 5:6].apply(LabelEncoder().fit_transform)
```

```python
data.iloc[:, 1:3] = data.iloc[:, 1:3].apply(LabelEncoder().fit_transform)
data.head()
```

```python
data = data.drop(['Number','Color','Egg_Group_1','Egg_Group_2','Body_Style','hasGender','Pr_Male','hasMegaEvolution','Height_m','Weight_kg'], axis = 1)
data.head()
```

```python
#assign target value
target = data['isLegendary']
#drop target value from the dataframe
data = data.drop('isLegendary', axis = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#create KNN object
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(data, target)
KNeighborsClassifier(algorithm = 'auto', leaf_size = 30,
                    metric = "minkowski", metric_params = None, n_jobs = None,
                    n_neighbors = 5, p = 2, weights = 'uniform')
```

```python
test_data = [136,143,670,34,72,145,100,135,95,2,127]
test_data = np.array(test_data).reshape(1, -1)
test_data
```

```python
prediction = knn.predict(test_data)
result = 'Legendary' if prediction == True else 'Not Legendary'
print(result)
```
